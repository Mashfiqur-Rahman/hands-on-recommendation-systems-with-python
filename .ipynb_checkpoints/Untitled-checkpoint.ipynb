```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
def euclidean(v1, v2):
    v1 = np.array(v1)
    v2 = np.array(v2)
    
    diff = np.power(np.array(v1) - np.array(v2), 2)
    sigma_val = np.sum(diff)
    euclid_score = np.sqrt(sigma_val)
    return euclid_score
```

```python
#Define 3 users with ratings for 5 movies
u1 = [5,1,2,4,5]
u2 = [1,5,4,2,1]
u3 = [5,2,2,4,4]
```

```python
euclidean(u1, u2)
```

```python
euclidean(u1, u3)
```

```python
alice = [1,1,3,2,4]
bob = [2,2,4,3,5]
```

```python
euclidean(alice, bob)
```

```python
eve = [5,5,3,4,2]
euclidean(eve, alice)
```

```python
from scipy.stats import pearsonr
```

```python
# The first element of our list output is the Pearson score
pearsonr(alice, bob)
```

```python
pearsonr(alice, eve)
```

```python
pearsonr(bob, eve)
```

```python
from sklearn.datasets import make_blobs
```

```python
X, y = make_blobs(n_samples=300, cluster_std=0.5, centers=3, random_state=0)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(X[:, 0], X[:, 1], s=50)
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=3, init='random', max_iter=10)
```

```python
kmeans.fit(X)
```

```python
y_pred = kmeans.predict(X)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y_pred, s=50)
```

```python
centroids = kmeans.cluster_centers_
```

```python
plt.subplots(figsize=(14,10))
plt.scatter(X[:, 0], X[:, 1],c=y_pred, s=50)
plt.scatter(centroids[:, 0], centroids[:, 1], c='black', s=100, marker='X')
```

```python
ss = []
for i in range(1, 9):
    kmeans = KMeans(n_clusters=i, init='random', max_iter=10, random_state=0).fit(X)
    ss.append(kmeans.inertia_)
    
import seaborn as sns
sns.pointplot(x=[j for j in range(1, 9)], y=ss)
```

```python
from sklearn.datasets import make_moons
```

```python
y_m_pred = kmm.predict(X_m)
```

```python
plt.scatter(X_m[:, 0], X_m[:, 1], c=y_m_pred, s=50)
```
